### some necessary functions:

```python
 def compute_loss(w):
    if loss_type=='coercive_saddle':
        loss=0.5*w[0]**2+1/4*w[1]**4-0.5*w[1]**2
    elif loss_type=='tube':
        d=w.shape[0]
        L=1.5
        gamma=1
        tau=1/2
        
        if w.all() >= 2*tau and 6*tau >= w.all():
            loss= np.dot(np.ones(d),L*(w-4*tau)**2)
        else:
            loss=0
            #1st dim
            if w[0] <=tau and w[1:d].all() <=tau and  w[1:d].all()>=0:
                loss=-gamma*w[0]**2+np.dot(np.ones(d-1),L*w[1:d]**2)
            elif w[0]>tau and w[0]<=2*tau and w[1:d].all() <=tau and  w[1:d].all()>=0:
                #loss=np.dot(np.ones(d-2),L*w[2:d]**2)
                loss=0
            #2nd to d-1th dim    
            for i in range(1,d-1): 
                if  w[0:i].all() >= 2*tau and w[0:i].all() <=6*tau and w[i:d].all() <=tau and  w[i:d].all()>=0:
                    _loss=np.dot(np.ones(i),L*(w[0:i]-4*tau)**2)-gamma*w[i]**2+np.dot(np.ones(d-(i+1),L*w[i+1:d]**2))
                    loss=loss+_loss
                elif w[0:i].all() >= 2*tau and w[0:i].all() <=6*tau and w[i] >tau and w[i]<=2*tau and w[i+1:d].all() <=tau and  w[i+1:d].all()>=0:
                    _loss=np.dot(np.ones(i),L*(w[0:i]-4*tau)**2)+np.dot(np.ones(d-(i+2),L*w[i+2:d]**2))
                    loss=loss+_loss
                    
            #d-th dim        
            if w[0:d-1].all() >= 2*tau and w[0:d-1].all() <=6*tau and w[d-1] <=tau and w[d-1] >=0:
                    loss=loss + np.dot(np.ones(d-1),L*(w[0:d-1]-4*tau)**2)-gamma*w[d-1]**2
            elif w[0:d-1].all() >= 2*tau and w[0:d-1].all() <=6*tau and w[d-1] > tau and w[d-1]<=2*tau:
                    loss=loss + np.dot(np.ones(d-1),L*(w[0:d-1]-4*tau)**2)                                                                                                                                                              
            
    else:
        raise ValueError("Unknown losstype!")
    
    return loss

def compute_gradients(w):
# forward pass

    if loss_type=='coercive_saddle':
        grad = np.array([w[0], w[1] ** 3 - w[1]])
    elif loss_type=='tube':
        L=1.5
        gamma=1
        tau=1/2
        d=w.shape[0]

        
        if w.all() >= 2*tau and 6*tau >= w.all():
            grad= 2*L*(w-4*tau)
        else:
            #1st to d-1st dim
            for i in range(1,d-1):
                if  w[0:i].all() >= 2*tau and w[0:i].all() <=6*tau and w[i:d].all() <=tau and  w[i:d].all()>=0:
                    _grad=np.array([-2*gamma*w[i],2*L*(w[0:i-1]-4*tau), 2*L*w[i+1:d]])
                    grad=grad+_grad
                elif w[0:i].all() >= 2*tau and w[0:i].all() <=6*tau and w[i] >tau and w[i]<=2*tau and w[i+1:d].all() <=tau and  w[i+1:d].all()>=0:
                    _grad=np.array([0,2*L*(w[0:i-1]-4*tau),2*L*w[i+2:d]])
                    grad=grad+_grad 
        
            #d-th dim        
            if w[0:d-1].all() >= 2*tau and w[0:d-1].all() <=6*tau and w[d-1] <=tau and w[d-1] >=0:  
                _grad=np.array([2*(w[0:d-1]-4*tau),-gamma*2*w[d-1]])
                grad=grad+_grad
            elif w[d-1] > tau and w[d-1]<=2*tau:
                _grad=np.array([2*(w[0:d-1]-4*tau),0])
                grad=grad+_grad            
    else:
        raise ValueError("Unknown losstype!")
        
    return grad



def make_plot(list_loss,list_x,list_methods,list_steps,limits):
    colors = ['#42A5F5', '#FFB300', '#BA68C8','#7CB342']

 #2. Plot performance over time/steps
    plt.figure(1)

    for i in range(len(list_loss)):
        plt.plot(list_x[i], list_loss[i],color= colors[i])
    plt.legend(list_methods,fontsize=12, loc=1)
   
    s=0.2
    x_ax = np.arange(limits[0],limits[1]+s, s)  
    y_ax = np.arange(limits[0],limits[1]+s, s)  
    x_ax,y_ax = np.meshgrid(x_ax,y_ax)
    loss=np.zeros((x_ax.shape[0],x_ax.shape[1]))

    for i in range(x_ax.shape[0]):
        for j in range(x_ax.shape[1]):
            loss[i,j]=compute_loss(np.array((x_ax[i,j],y_ax[i,j])))

    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax = fig.gca(projection='3d')
    ax.plot_surface(x_ax, y_ax, loss,cmap='viridis', 
                alpha=0.4,zorder=1)

```

### Gradient Descent

```python
# Run GD (simultaneously)
def Gradient_Descent(w,iterations,step_size,accuracy):
    iterates=[]
    loss=[]
    x=[]
    for k in range(iterations): 
        #safe information
        iterates.append(w)
        x.append(k-1)
        loss.append(compute_loss(w))
        
        #take step
        grad=compute_gradients(w)  
        w=w-step_size*grad
        if np.linalg.norm(grad)<accuracy:
            break

    return(iterates,x,loss)
```

### Accelerated Gradient Descent

```python
#Run AGD
def Accelerated_Gradient_Descent2(w,iterations,step_size, accuracy):
    iterates=[]
    loss=[]
    x=[]
    w_old=w

    for k in range(1,iterations+1): 
        iterates.append(w)
        x.append(k-1)
        loss.append(compute_loss(w))
        
        #compute y_(k)
        y=w+0.75*(w-w_old)
        if loss_type=='tube':
            if y[1]<0:
                y[1]=-y[1]
        #store x_(k-1)
        w_old=w
        #compute gradients
        grad=compute_gradients(y)
        w=y-step_size*grad
        if loss_type=='tube': #if no mirroring trick is used in gradients. apply here!
            if w[1]<0:
                print(' neg x_2')
            #    w[1]=-w[1]
 
        if np.linalg.norm(grad)<accuracy:

            break

    return(iterates,x,loss)
```

```python
#Run AGD
def Accelerated_Gradient_Descent(w,iterations,step_size, accuracy):
    iterates=[]
    loss=[]
    x=[]
    w_old=w

    for k in range(1,iterations+1): 
        iterates.append(w)
        x.append(k-1)
        loss.append(compute_loss(w))
        
        #compute y_(k)
        y=w+0.25*(w-w_old)
        if loss_type=='tube':
            if y[1]<0:
                y[1]=-y[1]
        #store x_(k-1)
        w_old=w
        #compute gradients
        grad=compute_gradients(y)
        w=y-step_size*grad
        if loss_type=='tube': #if no mirroring trick is used in gradients. apply here!
            if w[1]<0:
                print(' neg x_2')
            #    w[1]=-w[1]
 
        if np.linalg.norm(grad)<accuracy:

            break

    return(iterates,x,loss)
```

```python
def Perturbed_Accelerated_Gradient_Descent(w,iterations,step_size,accuracy,t_thres, g_thres,f_thres,radius):
    iterates=[]
    loss=[]
    x=[]
    w_old=w
  
    t_noise=-t_thres-1
    eta=step_size
    
    for k in range(iterations): 
        #safe information
        iterates.append(w)
        x.append(k-1)
        loss_w=compute_loss(w)
        loss.append(loss_w)
        
        #compute y_(k)
        y=w+(k-1)/(k+2)*(w-w_old)
        if loss_type=='tube':
            if y[1]<0:
                y[1]=-y[1]
        #store x_(k-1)
        w_old=w
        #compute gradients
        grad=compute_gradients(y) 
        
        #Should we perturb the extrapolation point or rather x_k+1.. My guess is the extrapolation point.
        
        if np.linalg.norm(grad)<g_thres:
            #Sample 1 point uniformly form ball with radius r 
            p=radius*np.ones(w_0.shape[0])
            while np.linalg.norm(p)>radius:
                p = radius*np.random.normal(size=(1, w_0.shape[0]))[0]
                if loss_type=='coercive_saddle':
                    if  p[1]>0: #make PGD turn left! (just for illustrative purposes) 
                        p[1]=-p[1]
                elif loss_type=='tube':
                    if  p[0]<0: #make sure PGD stays within [0,3]x[0,3]
                        p[0]=-p[0]
                    if  p[1]<0: 
                        p[1]=-p[1]
             #   print(k,'AGD perturbing')
            _y=y
            t_noise=k
            y=_y+p
            grad=compute_gradients(y)  
            
        w=y-step_size*grad
        
        if loss_type=='tube':
            if w[1]<0:
                print(' neg x_2')
                #w[1]=-w[1]
    
        if np.linalg.norm(grad)<accuracy:
            break

    return(iterates,x,loss)
```

### Momentum

```python
#Run Momentum
def Heavy_Ball(w,iterations,step_size, accuracy):
    iterates=[]
    loss=[]
    x=[]
    w_old=w

    for k in range(1,iterations+1): 
        iterates.append(w)
        x.append(k-1)
        loss.append(compute_loss(w))
        
        #compute y_(k)
        y=w+(k-1)/(k+2)*(w-w_old)
        #store x_(k-1)
        w_old=w
        #compute gradients
        grad=compute_gradients(w)
        w=y-step_size*grad
 
        if np.linalg.norm(grad)<accuracy:
            break

    return(iterates,x,loss)
```

### Pertubed Gradient Descent

```python
# Run GD (simultaneously)
def Pertubed_Gradient_Descent(w,iterations,step_size,accuracy,t_thres, g_thres,f_thres,radius):
    iterates=[]
    loss=[]
    x=[]
  
    t_noise=-t_thres-1
    eta=step_size
    
    for t in range(iterations): 
        #safe information
        iterates.append(w)
        x.append(t-1)
        loss_w=compute_loss(w)
        loss.append(loss_w)
        
        grad=compute_gradients(w)  
        
        if np.linalg.norm(grad)<g_thres:
            #Sample 1 point uniformly form ball with radius r 
            p=radius*np.ones(w_0.shape[0])
            while np.linalg.norm(p)>radius:
                p = radius*np.random.normal(size=(1, w_0.shape[0]))[0]
                if loss_type=='coercive_saddle':
                    if  p[1]>0: #make PGD turn left! (just for illustrative purposes) 
                        p[1]=-p[1]
                elif loss_type=='tube':
                    if  p[0]<0: #make sure PGD stays within [0,3]x[0,3]
                        p[0]=-p[0]
                    if  p[1]<0: 
                        p[1]=-p[1]
             #   print(t,'GD perturbing')

            _w=w
            t_noise=t
            w=_w+p
            grad=compute_gradients(w)              
            
        #if t-t_noise==t_thres and loss_w-compute_loss(_w)>-f_thres: #stop pertubing around minimizers
        #    iterates[-1]=_w
        #    loss[-1]=compute_loss(_w)
        #    break
        
        w=w-eta*grad
        if np.linalg.norm(grad)<accuracy:
            break

    return(iterates,x,loss)
```
